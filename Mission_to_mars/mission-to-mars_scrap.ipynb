```python
#import BeautifulSoup, Pandas, request/splinter

from bs4 import BeautifulSoup
import pandas as pd
import requests
from splinter import Browser
import pymongo 
```

```python
# https://splinter.readthedocs.io/en/latest/drivers/chrome.html
!which chromedriver
```

```python
# chromedriver/splinter
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

# NASA Mars News

```python
# NASA Mars News
#scrap the NASA mars news site https://mars.nasa.gov/news/

url_one = 'https://mars.nasa.gov/news/'
browser.visit(url_one)
```

```python
#create beautiful soup object
response = requests.get(url_one)

soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
# look at body to find div class, not pulling correctly when inspect site
para = soup.body
print(para)
```

```python
#collect the latest news title and paragraph text
#find('span', 'article_teaser_body')

news = soup.find('div', class_='content_title').text
print(f"The title is: {news}")
```

```python
para = soup.find('div', class_='rollover_description_inner').text
print(f"The teaser paragraph is: \n{para}")
```

# JPL Mars Space Images Featured Image

```python
# second URL to be scraped add browser.visit()
# visit the JPL site for the featured image (https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars

url_two = 'https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars'
browser.visit(url_two)
```

```python
# add html parser and create beautiful soup object
html = browser.html
soup_image = BeautifulSoup(html, 'html.parser')
```

```python
# click to next page to make image larger
browser.click_link_by_id('full_image')
```

```python
# more info button to make image bigger
# browser.click_link_by_partial_text('more info')

browser.links.find_by_partial_text('full_image')
```

```python
image_html = soup_image.body
print(image_html.prettify())
```

```python
# print image div class
image_path_two = soup_image.find("div", class_='carousel_items')
image_path_two
```

```python
# use splinter to nav and find image URL for current Featured Mars Image
# assign URL string to featured_image_url
# full size .jpg image and save url string

image_path = soup_image.find("div", class_='carousel_items')

link = image_path.find('style') ['background-image']

base_url = 'https://www.jpl.nasa.gov'

featured_image_url = base_url + link
print(featured_image_url)
```

```python
featured_image_path = soup_image.find('img')["src"]


print(featured_image_path)
```
